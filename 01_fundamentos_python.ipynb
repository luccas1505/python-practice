**CALCULADORA COM OPERADORES**

1-Crie um programa que solicite dois valores numéricos à pessoa usuária e imprima a soma dos dois valores.

```python
val1 = int(input("Digite o primeiro valor: "))
val2 = int(input("Digite o segundo valor: "))
valTot = val1 + val2

print(f"O valor final é {valTot}")
```

2-Crie um programa que solicite três valores numéricos à pessoa usuária e imprima a soma dos três valores.

```python
val1 = int(input("Digite o primeiro valor: "))
val2 = int(input("Digite o segundo valor: "))
val3 = int(input("Digite o terceiro valor: "))
valTot = val1+val2+val3

print(f"O valor total é {valTot}")
```

3-Crie um programa que solicite dois valores numéricos à pessoa usuária e imprima a subtração do primeiro pelo o segundo valor.

```python
val1 = int(input("Digite o primeiro valor: "))
val2 = int(input("Digite o segundo valor: "))
valTot = val1 - val2

print(f"O valor final é {valTot}")
```

4-Crie um programa que solicite dois valores numéricos à pessoa usuária e imprima a multiplicação dos dois valores.

```python
val1 = int(input("Digite o primeiro valor: "))
val2 = int(input("Digite o segundo valor: "))
valTot = val1 * val2

print(f"o valor total é de {valTot}")
```

5-Crie um código que solicita 3 notas de um estudante e imprima a média das notas.

```python
val1 = float(input("Digite a primeira nota: "))
val2 = float(input("Digite a segunda nota: "))
val3 = float(input("Digite a terceira nota: "))
valTot = (val1 + val2 + val3)/3

print(f"A média final das suas 3 notas é {valTot:.2f}")
```

10-Crie um código que calcule e imprima a média ponderada dos números 5, 12, 20 e 15 com pesos respectivamente iguais a 1, 2, 3 e 4.

```python
val1 = 5
val2 = 12
val3 = 20
val4 = 15
valTot = (((val1 *1) + (val2 *2) + (val3 * 3) + (val4 * 4))/10)

print(f"A nota final é de {valTot:.2f}")
```
